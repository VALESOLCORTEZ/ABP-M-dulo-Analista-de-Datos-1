¿Es lógico encontrar valores iguales a cero en determinados registros de esta variable? Al parecer en nuestro Dataframe no tenemos valores null, sin embargo vamos a ver que sucede si tratamos de rastrear los registros con valores "cero" y si guarda lógica con la variable utilizda.

Se observan muchas columnas con valores 0 lo cual tiene poco sentido practico en algunos casos, como por ejemplo en:

| Variable                            | ¿Puede valer 0?   | Comentario                                                                                         |
| ----------------------------------- | ----------------- | -------------------------------------------------------------------------------------------------- |
| Número de embarazos (NEMB)          | ✅ Sí              | Una persona puede no haber estado embarazada.                                                      |
| Glucosa en plasma (GLU)             | ❌ No lógico       | Glucosa 0 implicaría condición incompatible con la vida; suele ser dato faltante.                  |
| Presión arterial diastólica (PART)  | ❌ No lógico       | 0 mmHg no es real en una persona viva.                                                             |
| Grosor pliegue tríceps (GROS)       | ❌ Poco lógico     | 0 mm no es medición realista; suele indicar faltante.                                              |
| Insulina sérica 2h (HUR)            | ⚠️ Muy improbable | 0 puede aparecer por ausencia de medición o límite de detección. Suele tratarse como faltante.     |
| IMC (BMI)                           | ❌ No lógico       | IMC 0 imposible.                                                                                   |
| Función predicción diabetes (FPRED) | ⚠️ Podría ser 0   | Depende de cómo esté calculada; valores muy bajos son posibles, exactamente 0 depende del dataset. |
| Edad (AGE)                          | ❌ No lógico       | En este dataset de adultos, 0 imposible.                                                           |
| Clase (CLASS)                       | ✅ Sí              | 0 = no diabetes, 1 = diabetes.                                                                     |

### METODO FILLNA (RELLENAR con la media, mediana o algun numero en especifico)

El método fillna() de pandas sirve para reemplazar valores faltantes (NaN, None) en un DataFrame, en lugar de eliminar filas o columnas.

Es una forma de completar datos incompletos, cuando no se quiere perder registros usando dropna():

* media 👉 mean()
* mediana 👉 median()
* moda 👉 mode()
* 0 (cero) 👉 (0)
* texto específico 👉 ("Sin dato")
* valor anterior 👉 ffill()
* valor siguiente 👉 bfill()

El reemplazo podemos hacerlo:

**🎯 Con inplace=True**
* inplace=True significa que el método modifica directamente el DataFrame original, sin necesidad de guardar el resultado en otra variable.
* El original cambia directamente.
* in = dentro, place = lugar, 👉 “modificar en el mismo lugar”

**🎯 Sin inplace=True**
* El original queda igual.
* Guardamos la información en otra variable.


```python
# Podemos hacerlo directamente al Dataframe usando implace=(True)

# df.fillna(df.mean(), inplace=True)
# print(df.shape)
# print(df.head().round(2)) # round(2) sirve para redondear los números a 2 decimales.
```

```python
# Gurdando el Dataframe en una nueva variable (usando la media)
df_mean = df.fillna(df.mean()).round(2)
df_mean
```

```python
# Gurdando el Dataframe en una nueva variable (usando el valor anterior)
df_anterior = df.fillna(method="ffill").round(2) # tambien se puede escribir df_anterior = df.ffill().round(2)
df_anterior
```

```python
# Gurdando el Dataframe en una nueva variable (usando el valor siguiente)
df_posterior = df.fillna(method="bfill").round(2) # tambien se puede escribir df_posterior = df.bfill().round(2)
df_posterior
```

```python
# Gurdando el Dataframe en una nueva variable (usando el valor cero)
df_cero = df.fillna(0).round(2)
df_cero
```

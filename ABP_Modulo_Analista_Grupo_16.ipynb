# 2. Limpieza de Datos

```python
# Limpieza de Duplicados: Eliminamos registros repetidos que sesgan el análisis [
df_trafico = df_trafico.drop_duplicates()

# Veracidad de Fechas: Sincronización automática
# Corregimos el error del Excel donde el día no coincide con la fecha calendario
df_trafico['fecha'] = pd.to_datetime(df_trafico['fecha'], dayfirst=True)
dias_reales = {0: 'Lunes', 1: 'Martes', 2: 'Miércoles', 3: 'Jueves', 4: 'Viernes', 5: 'Sábado', 6: 'Domingo'}
df_trafico['dia_semana'] = df_trafico['fecha'].dt.dayofweek.map(dias_reales)

# Estandarización de la variable 'turno'
# Eliminamos el texto entre paréntesis (ej: "Mañana (6-12h)" -> "Mañana") para uniformidad
df_trafico['turno'] = df_trafico['turno'].str.split(' ').str[0]

# Normalización de Criterios: Limpieza de espacios en blanco
# Aplicamos strip() a todas las columnas de texto para evitar errores de categorías duplicadas
columnas_texto = df_trafico.select_dtypes(include=['object']).columns
for col in columnas_texto:
    df_trafico[col] = df_trafico[col].str.strip()

# Tratamiento de Valores Faltantes (Nulos)
# Eliminamos filas sin métricas críticas (flujo y velocidad) para no distorsionar la media
df_trafico = df_trafico.dropna(subset=['flujo_vehicular_veh_hora', 'velocidad_promedio_kmh'])

# Verificación de Formatos Numéricos
# Aseguramos que las métricas de tráfico sean tratadas como números (float)
cols_numericas = ['indice_congestion', 'demora_adicional_min', 'ocupacion_colectivo_pct']
df_trafico[cols_numericas] = df_trafico[cols_numericas].apply(pd.to_numeric, errors='coerce')

print("ETL: Limpieza, Veracidad y Estandarización Completada ")
print(f"Registros finales en df_trafico: {df_trafico.shape}")
df_trafico.head()
```

Extracción de variables

```python
# Definición de las 16 variables estratégicas para Smart City Córdoba
variables_elegidas = [
    'fecha', 'dia_semana', 'turno', 'corredor_vial', 'barrio',
    'flujo_vehicular_veh_hora', 'es_hora_pico', 'velocidad_promedio_kmh',
    'demora_adicional_min', 'indice_congestion', 'ciclo_semaforo_actual_seg',
    'ciclo_semaforo_optimo_seg', 'frecuencia_colectivo_min',
    'ocupacion_colectivo_pct', 'condicion_climatica', 'incidente_vial'
]

# Generación del dataframe oficial para el proyecto
# Se utiliza .copy() para asegurar que sea un objeto independiente del original
df_trafico_SmartCity = df_trafico[variables_elegidas].copy()

# Inspección técnica del nuevo dataframe
print(" Estructura de df_trafico_SmartCity ")
print(f"Total de registros: {df_trafico_SmartCity.shape}")
print(f"Total de variables: {df_trafico_SmartCity.shape[1]}")
print("\nResumen de columnas y tipos de datos:")
df_trafico_SmartCity.info()

# Visualización de los primeros registros
df_trafico_SmartCity.head()
```

```python
# Visualización de Variables Cualitativas (Modas)

# Configuración de estética para modo oscuro (Dark Mode)
plt.style.use('dark_background')
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('CBA Smart Mobility: Visualización de Patrones Críticos', color='#00FFCC', fontsize=22, weight='bold')

# 1. BARRIO: Gráfico de barras horizontales
# Usamos la paleta 'magma' y ordenamos por frecuencia para identificar la Moda visualmente [6, 7]
sns.countplot(data=df_trafico_SmartCity, y='barrio', ax=axes[0,0], palette='magma',
              order=df_trafico_SmartCity['barrio'].value_counts().index, hue='barrio', legend=False)
axes[0,0].set_title('Moda por Barrio (Frecuencia de Tráfico)', color='#FFD700', fontsize=15)
axes[0,0].set_xlabel('Cantidad de Registros', color='white')
axes[0,0].set_ylabel('')

# 2. ES HORA PICO: Gráfico circular (torta)
# Respetando los colores estratégicos: Rojo para Sí y Cian para No [7, 8]
data_pico = df_trafico_SmartCity['es_hora_pico'].value_counts()
axes[0,1].pie(data_pico, labels=data_pico.index, autopct='%1.1f%%',
               colors=['#FF4B4B', '#00FFCC'], startangle=90,
               textprops={'color':"white", 'weight': 'bold', 'fontsize': 12},
               wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})
axes[0,1].set_title('Proporción: Registros en Hora Pico', color='#FFD700', fontsize=15)

# 3. CONDICIÓN CLIMÁTICA: Gráfico de barras verticales
# Paleta 'flare' para diferenciar factores ambientales [7]
sns.countplot(data=df_trafico_SmartCity, x='condicion_climatica', ax=axes[1,0], palette='flare',
              order=df_trafico_SmartCity['condicion_climatica'].value_counts().index, hue='condicion_climatica', legend=False)
axes[1,0].set_title('Moda: Condición Climática', color='#FFD700', fontsize=15)
axes[1,0].set_xlabel('')
axes[1,0].set_ylabel('Cantidad', color='white')

# 4. INCIDENTE VIAL: Gráfico de barras verticales
# Paleta 'viridis' y rotación de etiquetas para mayor legibilidad [7]
sns.countplot(data=df_trafico_SmartCity, x='incidente_vial', ax=axes[1,1], palette='viridis',
              order=df_trafico_SmartCity['incidente_vial'].value_counts().index, hue='incidente_vial', legend=False)
axes[1,1].set_title('Moda: Tipos de Incidentes', color='#FFD700', fontsize=15)
axes[1,1].set_xlabel('')
axes[1,1].set_ylabel('Cantidad', color='white')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Cálculo numérico de la Moda para el informe técnico [4, 5]
cualitativas = ['barrio', 'condicion_climatica', 'es_hora_pico', 'incidente_vial']
resumen_cat = df_trafico_SmartCity[cualitativas].mode()
print("\n Resumen Estadístico: Modas de Variables Cualitativas")
print(resumen_cat)
```

Variables cuantitativas

```python
# Visualización de Variables Cuantitativas (Distribución y Dispersión)

# Configuración de estética para modo oscuro
plt.style.use('dark_background')
fig, axes = plt.subplots(2, 4, figsize=(22, 10))
fig.suptitle('CBA Smart Mobility: Análisis de Distribución y Outliers', color='#00FFCC', fontsize=22, weight='bold')

# Lista de las 4 variables numéricas core del proyecto
cuantitativas = ['flujo_vehicular_veh_hora', 'velocidad_promedio_kmh', 'indice_congestion', 'demora_adicional_min']

# Generación automática de Histogramas (Fila 1) y Boxplots (Fila 2)
for i, var in enumerate(cuantitativas):
    # 1. HISTOGRAMAS: Muestran la frecuencia y la forma de la distribución (Asimetría)
    sns.histplot(df_trafico_SmartCity[var], kde=True, ax=axes[0, i], color='#00FFCC', bins=20)
    axes[0, i].set_title(f'Distribución: {var}', color='#FFD700', fontsize=14)
    axes[0, i].set_xlabel('')
    axes[0, i].set_ylabel('Frecuencia', color='white')

    # 2. BOXPLOTS: Ideales para ver la Mediana, el Rango Intercuartílico y Outliers
    sns.boxplot(x=df_trafico_SmartCity[var], ax=axes[1, i], color='#FF4B4B',
                flierprops={'markerfacecolor': '#FFD700', 'marker': 'o'})
    axes[1, i].set_title(f'Outliers/Dispersión: {var}', color='#FFD700', fontsize=14)
    axes[1, i].set_xlabel('')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Resumen numérico de las medidas de la Unidad 1 para estas variables
print("\n Medidas de Tendencia Central y Dispersión (Resumen Técnico)")
resumen_num = df_trafico_SmartCity[cuantitativas].describe().T
resumen_num['varianza'] = df_trafico_SmartCity[cuantitativas].var()
print(resumen_num[['mean', '50%', 'std', 'varianza', 'min', 'max']])
```

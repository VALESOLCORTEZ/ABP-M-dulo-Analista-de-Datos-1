# Inyección de ruido controlado en dataset formato csv

Como parte de la preparación del dataset de tráfico para la etapa de ETL, se aplicó un proceso de inyección de anomalías diseñado para simular el desgaste natural de los datos en entornos operativos reales. La metodología evolucionó desde la inserción inicial de errores estáticos y proporciones simétricas —que generaban sesgos artificiales y colapsaban los cálculos de dispersión— hacia un modelo de caos asimétrico fundamentado en la detección dinámica de variables. En su versión final, el proceso introduce de manera aleatoria e impredecible entre un 2% y un 6% de registros duplicados para simular fallos de sincronización de red, e inyecta proporciones asimétricas de valores nulos (del 1% al 12%) imitando micro-cortes en los sensores. Además, se generan valores atípicos alterando entre el 1% y el 4% de las métricas numéricas mediante multiplicadores lógicos para preservar la viabilidad del análisis estadístico, y se simula el error humano en la carga manual afectando del 1% al 5% de los campos de texto con espacios invisibles o desajustes de mayúsculas. Este enfoque garantiza un nivel de "suciedad realista", proporcionando un entorno óptimo para poner a prueba la robustez de las transformaciones y técnicas de limpieza posteriores.

```python

import random
import math

# PARTE 1: INYECCIÓN DE ERRORES ASIMÉTRICOS (Caos Realista)

df_sucio = df.copy()

cols_numericas = df_sucio.select_dtypes(include=['number']).columns.tolist()
cols_texto = df_sucio.select_dtypes(include=['object']).columns.tolist()

def elegir_multiples(lista, cant):
    return random.sample(lista, min(cant, len(lista))) if lista else []

# 1. Duplicados (Aleatorio entre 2% y 6%)
num_dup = int(len(df_sucio) * random.uniform(0.02, 0.06))
df_sucio = pd.concat([df_sucio, df_sucio.sample(n=num_dup, random_state=42)], ignore_index=True)

# 2. Nulos Asimétricos (Porcentajes distintos para cada columna, entre 1% y 12%)
for col in elegir_multiples(cols_numericas, 3) + elegir_multiples(cols_texto, 3):
    porcentaje_azar = random.uniform(0.01, 0.12) # Cada columna recibe un daño distinto
    df_sucio.loc[df_sucio.sample(frac=porcentaje_azar).index, col] = np.nan

# 3. Atípicos (Daño aleatorio entre 1% y 4%, con multiplicadores distintos)
for col in elegir_multiples(cols_numericas, 3):
    idx = df_sucio.sample(frac=random.uniform(0.01, 0.04)).index
    # A veces multiplica por 10, a veces vuelve el número negativo
    df_sucio.loc[idx, col] = df_sucio.loc[idx, col] * random.choice([10, -1, 20])

# 4. Errores de Tipeo (Cantidades aleatorias)
cols_tipeo = elegir_multiples(cols_texto, 3)
if len(cols_tipeo) > 0:
    df_sucio.loc[df_sucio.sample(frac=random.uniform(0.01, 0.05)).index, cols_tipeo[0]] = df_sucio[cols_tipeo[0]].astype(str) + " "
if len(cols_tipeo) > 1:
    df_sucio.loc[df_sucio.sample(frac=random.uniform(0.01, 0.05)).index, cols_tipeo[1]] = df_sucio[cols_tipeo[1]].astype(str).str.lower()
if len(cols_tipeo) > 2:
    df_sucio.loc[df_sucio.sample(frac=random.uniform(0.01, 0.05)).index, cols_tipeo[2]] = "-" + df_sucio[cols_tipeo[2]].astype(str)

df_sucio = df_sucio.sample(frac=1, random_state=7).reset_index(drop=True)
df_sucio.to_csv("Dataset_Trafico_Cordoba_Inyeccion_Perfecta.csv", index=False)
print("✅ Inyección completada. Archivo guardado.\n")
```

# 2. Limpieza de Datos

E.T.L.

```python
# LIMPIEZA DE DATOS (ETL) Modelo

print("Iniciando proceso de limpieza de datos (ETL)...")

df_limpio = df_sucio.copy()

# 1. Normalización de nombres de columnas
nuevas_columnas = []
for col in df_limpio.columns:
    # Elimina espacios en los extremos, reemplaza espacios internos y capitaliza la primera letra
    col_norm = col.strip().replace(" ", "_").capitalize()
    nuevas_columnas.append(col_norm)

df_limpio.columns = nuevas_columnas
print("✓ Columnas normalizadas")


# 2. Normalización de estructuras de datos (Fechas y Números)
# Aseguramos que 'Fecha' sea tratada como un objeto datetime real y no como texto
if 'Fecha' in df_limpio.columns:
    df_limpio['Fecha'] = pd.to_datetime(df_limpio['Fecha'], format='%d/%m/%Y', errors='coerce')

# Forzamos la estructura numérica en variables clave por si se leyeron como object
cols_numericas = ['Velocidad_promedio_kmh', 'Demora_adicional_min', 'Temperatura_c', 'Indice_congestion', 'Frecuencia_colectivo_min']
for col in cols_numericas:
    if col in df_limpio.columns:
        df_limpio[col] = pd.to_numeric(df_limpio[col], errors='coerce')
print("✓ Estructuras de datos (fechas y números) normalizadas.")



# 3. Tratamiento de duplicados y nulos
pasos_etl = ['tratamiento_duplicados', 'tratamiento_nulos']

for paso in pasos_etl:
    if paso == 'tratamiento_duplicados':
        filas_antes = len(df_limpio)
        df_limpio = df_limpio.drop_duplicates()
        print(f"✓ Duplicados: Se eliminaron {filas_antes - len(df_limpio)} filas redundantes.")

    elif paso == 'tratamiento_nulos':
        # Iteramos por cada columna para hacer una imputación inteligente según el tipo de dato
        for col in df_limpio.columns:
            if df_limpio[col].isnull().sum() > 0:
                # Si es numérica, usamos la mediana
                if pd.api.types.is_numeric_dtype(df_limpio[col]):
                    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].median())
                # Si es categórica/texto, usamos la moda
                else:
                    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].mode()[0])
        print("✓ Nulos: Valores faltantes imputados iterativamente por columna.")


# 4. Corrección de inconsistencias, strings y normalización de la columna Turno
if 'Condicion_climatica' in df_limpio.columns:
    df_limpio['Condicion_climatica'] = df_limpio['Condicion_climatica'].replace({'Nblado': 'Nublado'})

if 'Dia_semana' in df_limpio.columns:
    df_limpio['Dia_semana'] = df_limpio['Dia_semana'].str.title()

if 'Turno' in df_limpio.columns:
    mapeo_turnos = {
        'Mañana (6-12h)': 'mañana',
        'Tarde (12-18h)': 'tarde',
        'Noche (18-24h)': 'noche',
        'Madrugada (0-6h)': 'madrugada'
    }
    df_limpio['Turno'] = df_limpio['Turno'].replace(mapeo_turnos)

# Eliminar espacios sobrantes en todas las columnas de texto
for col in df_limpio.select_dtypes(include='object').columns:
    df_limpio[col] = df_limpio[col].astype(str).str.strip()

# Corregir valores de Turno que quedaron con "-"
if 'Turno' in df_limpio.columns:
    df_limpio['Turno'] = df_limpio['Turno'].replace({
        '-Mañana (6-12h)': 'mañana',
        '-Tarde (12-18h)': 'tarde',
        '-Noche (18-24h)': 'noche',
        '-Madrugada (0-6h)': 'madrugada'
    })

# Normalizar nuevamente Turno por seguridad
if 'Turno' in df_limpio.columns:
    mapeo_turnos = {
        'Mañana (6-12h)': 'mañana',
        'Tarde (12-18h)': 'tarde',
        'Noche (18-24h)': 'noche',
        'Madrugada (0-6h)': 'madrugada'
    }
    df_limpio['Turno'] = df_limpio['Turno'].replace(mapeo_turnos)
print("✓ Strings e inconsistencias de texto corregidas.")


# 5. Tratamiento de outliers extremos
print("Neutralizando valores atípicos...")

# Iteramos sobre las columnas numéricas que definiste antes
for col in cols_numericas:
    if col in df_limpio.columns:
        # Calculamos los cuartiles y el Rango Intercuartílico (IQR)
        Q1 = df_limpio[col].quantile(0.25)
        Q3 = df_limpio[col].quantile(0.75)
        IQR = Q3 - Q1

        # Definimos los límites lógicos superior e inferior
        lim_inf = Q1 - 1.5 * IQR
        lim_sup = Q3 + 1.5 * IQR

        # Topamos los valores que se salen de los límites
        df_limpio[col] = np.where(df_limpio[col] < lim_inf, lim_inf, df_limpio[col])
        df_limpio[col] = np.where(df_limpio[col] > lim_sup, lim_sup, df_limpio[col])

print("✓ Valores atípicos neutralizados.")


print("\n RESUMEN POST-LIMPIEZA")
df_limpio.info()
```

ANÁLISIS EXPLORATORIO ORIENTADO A LAS HIPÓTESIS DE INVESTIGACIÓN

```python
# 1. IMPACTO DEL CLIMA EN LA VELOCIDAD

print("1. IMPACTO DEL CLIMA EN LA VELOCIDAD")

impacto_clima = (
    df_analisis_Final
    .groupby('Condicion_climatica')['Velocidad_promedio_kmh']
    .agg(['mean', 'median', 'count'])
    .round(2)
)

impacto_clima.columns = [
    'Velocidad Media (km/h)',
    'Velocidad Mediana (km/h)',
    'Cantidad de Registros'
]

display(
    impacto_clima.sort_values(
        by='Velocidad Media (km/h)',
        ascending=False
    )
)

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df_analisis_Final,
    x='Condicion_climatica',
    y='Velocidad_promedio_kmh',
    palette="magma"
)

plt.title(
    'Velocidad Promedio según Condición Climática',
    fontsize=16,
    weight='bold'
)

plt.xlabel('Condición Climática')
plt.ylabel('Velocidad Promedio (km/h)')
plt.tight_layout()
plt.show()


# 2. BARRIOS CON MAYOR CONGESTIÓN

print("\n2. ANÁLISIS DE CONGESTIÓN POR BARRIO")

congestion_barrio = pd.crosstab(
    df_analisis_Final['Barrio'],
    df_analisis_Final['Nivel_congestion']
)

display(
    congestion_barrio.sort_values(
        by='Crítico',
        ascending=False
    )
)

plt.figure(figsize=(12,6))

sns.heatmap(
    congestion_barrio,
    annot=True,
    cmap='mako',
    fmt='g',
    linewidths=0.5
)

plt.title(
    'Nivel de Congestión según Barrio',
    fontsize=16,
    weight='bold'
)

plt.ylabel('Barrio')
plt.xlabel('Nivel de Congestión')
plt.tight_layout()
plt.show()


# 3. DEMORA SEGÚN TIPO DE INCIDENTE


print("\n3. IMPACTO DE LOS INCIDENTES VIALES EN LA DEMORA")

demora_incidentes = (
    df_analisis_Final
    .groupby('Incidente_vial')['Demora_adicional_min']
    .agg(['mean', 'median', 'count'])
    .round(2)
)

demora_incidentes.columns = [
    'Demora Media (min)',
    'Demora Mediana (min)',
    'Cantidad de Registros'
]

display(
    demora_incidentes.sort_values(
        by='Demora Media (min)',
        ascending=False
    )
)

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df_analisis_Final,
    x='Incidente_vial',
    y='Demora_adicional_min',
    palette='viridis',
    legend=False
)

plt.title(
    'Demora Adicional según Tipo de Incidente',
    fontsize=16,
    weight='bold'
)

plt.xlabel('Tipo de Incidente')
plt.ylabel('Demora Adicional (min)')
plt.xticks(rotation=20)

plt.tight_layout()
plt.show()


# 4. ANÁLISIS POR BARRIO Y DEMORA


print("\n4. BARRIOS CON MAYOR DEMORA PROMEDIO")

analisis_barrio = (
    df_analisis_Final
    .groupby('Barrio')
    .agg(
        Demora_Media_min=('Demora_adicional_min', 'mean'),
        Demora_Mediana_min=('Demora_adicional_min', 'median'),
        Velocidad_Media_kmh=('Velocidad_promedio_kmh', 'mean')
    )
    .round(2)
)

display(
    analisis_barrio.sort_values(
        by='Demora_Media_min',
        ascending=False
    )
)
```

### **Conclusiones Generales e Interpretación Analítica**

El presente análisis exploratorio permitió caracterizar el comportamiento del tránsito urbano de la ciudad de Córdoba a partir de un conjunto de 612 registros depurados mediante procesos de limpieza, normalización e imputación de datos. La aplicación de técnicas ETL posibilitó eliminar inconsistencias, corregir valores atípicos y garantizar la calidad de la información utilizada en el estudio.

- Los resultados obtenidos muestran que la velocidad promedio de circulación fue de 31,69 km/h, mientras que la mediana alcanzó los 33,6 km/h. La diferencia entre ambas medidas sugiere la existencia de episodios de circulación particularmente lenta que afectan el promedio general. Asimismo, la demora adicional promedio registrada fue de 1,72 minutos por recorrido, con una mediana de 1,3 minutos y un máximo de 4,3 minutos, reflejando que la mayoría de los desplazamientos presentan retrasos moderados, aunque existen situaciones puntuales de mayor impacto.

- Respecto de la primera pregunta de investigación, relacionada con el efecto de las condiciones climáticas sobre la velocidad de circulación, se observó que la velocidad media varió entre 30,91 km/h y 32,14 km/h según el estado del tiempo. Las velocidades más bajas se registraron durante episodios de llovizna, mientras que los valores más elevados se observaron en jornadas de lluvia y cielo nublado. Aunque las diferencias encontradas no fueron extremas, los resultados permiten inferir que las condiciones climáticas influyen en el comportamiento del tránsito urbano.

- En relación con la congestión vehicular, el análisis evidenció que el 35,8% de los registros correspondió a niveles bajos de congestión, el 36,6% a niveles moderados, el 14,7% a niveles altos y el 12,9% a niveles críticos. Esto implica que aproximadamente uno de cada cuatro eventos analizados presentó condiciones severas de circulación, constituyendo un indicador relevante para la gestión de la movilidad urbana.

- Por otra parte, el análisis territorial permitió identificar diferencias significativas entre barrios. Los cruces entre barrio y nivel de congestión mostraron que determinados sectores concentran una mayor cantidad de eventos de congestión elevada, lo que confirma que el fenómeno no se distribuye de manera homogénea dentro de la ciudad y que existen zonas particularmente vulnerables a los embotellamientos.

Finalmente, el estudio de las demoras asociadas a incidentes viales permitió observar que la presencia de accidentes u obras en la vía incrementa los tiempos de desplazamiento respecto de las situaciones sin incidentes registrados, evidenciando el impacto que estos eventos tienen sobre la fluidez del tránsito.

En conjunto, los resultados obtenidos permiten transformar datos operativos de circulación vehicular en información estadísticamente fundamentada para comprender mejor la dinámica del tránsito urbano. El análisis desarrollado constituye una base sólida para futuros estudios predictivos y para la formulación de estrategias orientadas a optimizar la movilidad y reducir los niveles de congestión en la ciudad de Córdoba.
